```python
#frame extractions:
import sys
import argparse

import cv2

def extract_images(pathIn, pathOut):
    count = 0
    vidcap = cv2.VideoCapture(pathIn)
    success,image = vidcap.read()
    success = True
    try:
      os.mkdir(pathOut)
    except FileExistsError:
      print("directory already exists")
    try:
      while success:
          vidcap.set(cv2.CAP_PROP_POS_MSEC,(count*1000))    # added this line
          success,image = vidcap.read()
          print ('Read a new frame: ', success)
          cv2.imwrite( pathOut + "/frame%d.jpg" % count, image)     # save frame as JPEG file
          count = count + 1
    except:
      print("end of video")
```

```python
import numpy as np
#extracting the bounding box coordinates
def bounding_box_fun(model_name:str,image_path:str,semantic_model:str)->list:
  model = YOLO(model_name)
  #image = cv2.imread(image_path)
  results = model(image_path)
  list_cord = []
  if semantic_model == "sam2":
    for coor_obj in results[0].boxes:
      list_cord.append(coor_obj.xyxy.cpu()[0])
  else:
    for coor_obj in results[0].boxes:
      list_cord.append(coor_obj.xywh.cpu()[0])
  return np.array(list_cord)



```

```python
import cv2
import numpy as np

# Lista dei file video da unire
video_files = ['/content/demo_cane.mp4', '/content/demo_cane.mp4']  # Aggiungi quanti ne vuoi

# Aprire tutti i video
caps = [cv2.VideoCapture(f) for f in video_files]

# Verifica che tutti i video siano stati caricati correttamente
if not all(cap.isOpened() for cap in caps):
    print("Errore nell'apertura di uno o più video.")
    exit()

# Ottenere parametri video (usiamo il primo video come riferimento)
frame_width = int(caps[0].get(3))
frame_height = int(caps[0].get(4))
fps = int(caps[0].get(cv2.CAP_PROP_FPS))
fourcc = cv2.VideoWriter_fourcc(*'mp4v')

# Creare il VideoWriter per salvare il nuovo video
out = cv2.VideoWriter('output.mp4', fourcc, fps, (frame_width, frame_height))

while True:
    frames = []
    ret_flags = []

    # Leggere un frame da ogni video
    for cap in caps:
        ret, frame = cap.read()
        frames.append(frame if ret else None)
        ret_flags.append(ret)

    # Se tutti i video sono terminati, esci dal loop
    if not any(ret_flags):
        break

    # Gestire video di lunghezze diverse: se un video è finito, usiamo l'ultimo frame disponibile
    for i in range(len(frames)):
        if frames[i] is None:
            frames[i] = frames[i - 1] if i > 0 else np.zeros((frame_height, frame_width, 3), dtype=np.uint8)

    # Calcolare il frame medio (blending di tutti i video)
    blended_frame = np.mean(frames, axis=0).astype(np.uint8)

    # Scrivere il frame nel nuovo video
    out.write(blended_frame)

    # Mostrare il video combinato (opzionale)
    #cv2.imshow('Combined Tracking', blended_frame)
    #if cv2.waitKey(1) & 0xFF == ord('q'):
    #    break

# Rilasciare risorse
for cap in caps:
    cap.release()
out.release()
cv2.destroyAllWindows()

```

```python
#merged tracked videos
import cv2
from google.colab.patches import cv2_imshow

cap1 = cv2.VideoCapture('/content/demo.mp4')
cap2 = cv2.VideoCapture('/content/demo_cane.mp4')

# Ottenere le caratteristiche del video
frame_width = int(cap1.get(3))  # Larghezza
frame_height = int(cap1.get(4))  # Altezza
fps = int(cap1.get(cv2.CAP_PROP_FPS))  # FPS
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec MP4

# Creare il VideoWriter per salvare il nuovo video
out = cv2.VideoWriter('/content/output.mp4', fourcc, fps, (frame_width, frame_height))

while cap1.isOpened() and cap2.isOpened():
    ret1, frame1 = cap1.read()
    ret2, frame2 = cap2.read()
    if not ret1 or not ret2:
        break

    # Unione dei due video
    blended = cv2.addWeighted(frame1, 0.5, frame2, 0.5, 0)

    # Salvare il frame nel nuovo video
    out.write(blended)

    # Mostra il video combinato
    #cv2_imshow( blended)
    #if cv2.waitKey(30) & 0xFF == ord('q'):
    #    break

cap1.release()
cap2.release()
cv2.destroyAllWindows()
```
